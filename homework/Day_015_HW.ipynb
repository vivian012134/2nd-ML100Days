```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

1. 請用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap

```python
matrix = []

for i in range(0,10):
    ran = np.random.uniform(-1, 1, 10)
    matrix.append(ran)

n_matrix = pd.DataFrame(matrix)

plt.figure(figsize=(10,10))
heatmap = sns.heatmap(n_matrix, cmap = plt.cm.RdYlBu_r, vmin = -1, annot = True, vmax = 1)
plt.title('Teat Heatmap');

plt.show()
```

2. 請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3
matrix = []

for i in range(0,ncol):
    ran = np.random.uniform(-1, 1, nrow)
    matrix.append(ran)

n_matrix = pd.DataFrame(matrix)

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(n_matrix, indice)

grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

3. 請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3
matrix = []

for i in range(0,ncol):
    ran = np.random.normal(0, 1, nrow)
    matrix.append(ran)
    
n_matrix = pd.DataFrame(matrix)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(n_matrix, indice)

grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(plt.scatter, alpha = 0.2)

grid.map_diag(sns.kdeplot)

grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
